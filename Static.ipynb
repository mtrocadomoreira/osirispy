```python
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.colors import LogNorm
import matplotlib.gridspec as gridspec # GRIDSPEC !
from matplotlib.ticker import FormatStrFormatter
ratio=16/9

plt.style.use('default')
mpl.rcParams['text.usetex'] = True
mpl.rcParams['text.latex.preamble'] = r'\usepackage{amsmath}' #for \text command
fig=plt.figure(figsize=(16,9),dpi=120)
plt.rcParams.update({'font.size': 38})
mpl.rc('font', family = 'serif')
plt.axes([0.0,0,1,1])

gs = gridspec.GridSpec(1,2, height_ratios=[1], width_ratios=[0.96,0.04],wspace=0.01)
gs.update(left=0.05, right=0.95, bottom=0.08, top=0.93, wspace=0.02, hspace=0.03)


ax = plt.subplot(gs[0,0]) # place it where it should be.
#ax=plt.gca()


N=42
xmin=0
xmax=40
ymin=0
ymax=20
extent=[xmin,xmax,ymin,ymax]
im=ax.imshow(np.random.random((N,N)),cmap="afmhot",aspect='auto',origin="lower",extent=extent)
#set_size(16,16/ratio) 

from mpl_toolkits.axes_grid1 import make_axes_locatable
l, b, w, h = ax.get_position().bounds
color_axis  = plt.subplot(gs[0,1]) # place it where it should be.

#the_divider = make_axes_locatable(ax)
#color_axis = the_divider.append_axes("right", size="5%", pad=0.0)

cbar=plt.colorbar(im,cax=color_axis)
cbar.set_label(r"$|E_2|^2 \;[m_e\text{c}\omega_p/e]$")
cbar.ax.yaxis.set_major_formatter(FormatStrFormatter('%.2f'))

ax.set_xlabel(r"$t_{det}~[\omega_p^{-1}]$")
ax.set_ylabel(r"$\phi\;$[mrad]")
xlen=ax.get_xlim()[1]-ax.get_xlim()[0]
ylen=ax.get_ylim()[1]-ax.get_ylim()[0]
ax.set_aspect(xlen/ylen/ratio)
#ax.yaxis.set_major_formatter(FormatStrFormatter('%.1f'))



plt.savefig("test3.pdf",facecolor=(0.0, 0.0, 0.0, 0.0),bbox_inches="tight")
```

```python
#transparency colormaps
from matplotlib.colors import ListedColormap
import matplotlib.pylab as pl

# Choose colormap
cmap = pl.cm.jet_r

# Get the colormap colors
my_cmap = cmap(np.arange(cmap.N))

# Set alpha
my_cmap[:,-1] = np.linspace(0, 1, cmap.N)

# Create new colormap
jet_r_t = ListedColormap(my_cmap)


import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.colors import LogNorm
import matplotlib.gridspec as gridspec # GRIDSPEC !
from matplotlib.ticker import FormatStrFormatter

ratio=16/9

plt.style.use('default')
mpl.rcParams['text.usetex'] = True
mpl.rcParams['text.latex.preamble'] = r'\usepackage{amsmath}' #for \text command
fig=plt.figure(figsize=(9*ratio*1.216,9),dpi=120)
plt.rcParams.update({'font.size': 38})
mpl.rc('font', family = 'serif')
plt.axes([0.0,0,1,1])



gs = gridspec.GridSpec(1,4, height_ratios=[1], width_ratios=[0.82,0.04,0.1,0.04],wspace=0.01)
gs.update(left=0.05, right=0.95, bottom=0.08, top=0.93, wspace=0.02, hspace=0.03)


ax = plt.subplot(gs[0,0]) # place it where it should be.
#ax=plt.gca()


N=42
xmin=0
xmax=40
ymin=0
ymax=20
extent=[xmin,xmax,ymin,ymax]
im=ax.imshow(np.random.random((N,N)),cmap="afmhot",aspect='auto',origin="lower",extent=extent)
#set_size(16,16/ratio) 

from mpl_toolkits.axes_grid1 import make_axes_locatable
l, b, w, h = ax.get_position().bounds
color_axis  = plt.subplot(gs[0,1]) # place it where it should be.

#the_divider = make_axes_locatable(ax)
#color_axis = the_divider.append_axes("right", size="5%", pad=0.0)

cbar=plt.colorbar(im,cax=color_axis)
cbar.set_label(r"$|E_2|^2 \;[m_e\text{c}\omega_p/e]$",labelpad=-112,y=0.78,va="top")
cbar.ax.yaxis.set_major_formatter(FormatStrFormatter('%.2f'))



def func3(x, y):
    return  np.exp(-((x-10)**2 + (y-10)**2)/16)


x = np.arange(0.0, 40.0, 0.1)
y = np.arange(0.0, 20.0, 0.1)
X, Y = np.meshgrid(x, y)
Z2 = func3(X, Y)

im2 = ax.imshow(Z2, cmap=jet_r_t,  interpolation='bilinear', origin="lower",   extent=extent)

color_axis2  = plt.subplot(gs[0,3]) # place it where it should be.

cbar=plt.colorbar(im2,cax=color_axis2)
cbar.set_label(r"$|E_2|^2 \;[m_e\text{c}\omega_p/e]$",labelpad=-112,y=0.78,va="top")
cbar.ax.yaxis.set_major_formatter(FormatStrFormatter('%.2f'))



ax.set_xlabel(r"$t_{det}~[\omega_p^{-1}]$")
ax.set_ylabel(r"$\phi\;$[mrad]")
xlen=ax.get_xlim()[1]-ax.get_xlim()[0]
ylen=ax.get_ylim()[1]-ax.get_ylim()[0]
ax.set_aspect(xlen/ylen/ratio)
#ax.yaxis.set_major_formatter(FormatStrFormatter('%.1f'))



plt.savefig("test5.pdf",facecolor=(0.0, 0.0, 0.0, 0.0),bbox_inches="tight")
```

### 16:9 Line plot

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.colors import LogNorm

ratio=16/9

plt.style.use('default')
mpl.rcParams['text.usetex'] = True
mpl.rcParams['text.latex.preamble'] = r'\usepackage{amsmath}' #for \text command
fig=plt.figure(figsize=(9*ratio,9),dpi=120)
plt.rcParams.update({'font.size': 38})
mpl.rc('font', family = 'serif')
plt.axes([0.0,0,1,1])



gs = gridspec.GridSpec(1,2, height_ratios=[1], width_ratios=[0.95,0.05],wspace=0.01)
gs.update(left=0.05, right=0.95, bottom=0.08, top=0.93, wspace=0.02, hspace=0.03)


ax = plt.subplot(gs[0,0]) # place it where it should be.

N=42
xmin=0
xmax=40
ymin=0
ymax=20
x=np.linspace(0,10)
ax.plot(x,x*x,label="Label")

ax.set_xlabel(r"$t_{det}~[\omega_p^{-1}]$")
ax.set_ylabel(r"$E_2\;[m_e\text{c}\omega_p/e]$")
xlen=ax.get_xlim()[1]-ax.get_xlim()[0]
ylen=ax.get_ylim()[1]-ax.get_ylim()[0]
ax.set_aspect(xlen/ylen/ratio)
ax.grid(ls="--",alpha=0.5)
ax.legend()
#plt.savefig("test4.pdf",facecolor=(0.0, 0.0, 0.0, 0.0),bbox_inches="tight")
```

### 16:9 Scatter Plot

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.colors import LogNorm

ratio=16/9

plt.style.use('default')
mpl.rcParams['text.usetex'] = True
mpl.rcParams['text.latex.preamble'] = r'\usepackage{amsmath}' #for \text command
fig=plt.figure(figsize=(9*ratio,9),dpi=120)
plt.rcParams.update({'font.size': 38})
mpl.rc('font', family = 'serif')
plt.axes([0.0,0,1,1])



gs = gridspec.GridSpec(1,2, height_ratios=[1], width_ratios=[0.95,0.05],wspace=0.01)
gs.update(left=0.05, right=0.95, bottom=0.08, top=0.93, wspace=0.02, hspace=0.03)


ax = plt.subplot(gs[0,0]) # place it where it should be.

N=42
xmin=0
xmax=40
ymin=0
ymax=20
x = np.random.randn(1000)
y = np.random.randn(1000)

ax.scatter(x,y,alpha=0.4,label="Label")

ax.set_xlabel(r"$t_{det}~[\omega_p^{-1}]$")
ax.set_ylabel(r"$E_2\;[m_e\text{c}\omega_p/e]$")
ax.set_xlim(-4,4)
ax.set_ylim(-4,4)
xlen=ax.get_xlim()[1]-ax.get_xlim()[0]
ylen=ax.get_ylim()[1]-ax.get_ylim()[0]
ax.set_aspect(xlen/ylen/ratio)
ax.grid(ls="--",alpha=0.5)
ax.legend()
plt.savefig("testscat.pdf",facecolor=(0.0, 0.0, 0.0, 0.0),bbox_inches="tight")
```
