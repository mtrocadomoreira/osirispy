```python
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.colors import LogNorm

plt.style.use('default')
mpl.rcParams['text.usetex'] = True
mpl.rcParams['text.latex.preamble'] = r'\usepackage{amsmath}' #for \text command
fig=plt.figure(figsize=(16,9),dpi=80)
plt.rcParams.update({'font.size': 32})
mpl.rc('font', family = 'serif')
ax=plt.gca()


N=42
xmin=0
xmax=20
ymin=0
ymax=10
extent=[xmin,xmax,ymin,ymax]
im=ax.imshow(np.random.random((N,N)),cmap="afmhot",aspect='auto',origin="lower",extent=extent)
             


#ax.set_aspect('equal')
from mpl_toolkits.axes_grid1 import make_axes_locatable

the_divider = make_axes_locatable(ax)
color_axis = the_divider.append_axes("right", size="5%")#, pad=0.1)
cbar=plt.colorbar(im,pad=2,cax=color_axis)
cbar.set_label(r"$|E_2|^2 \;[m_ec\omega_p/e]$",labelpad=-120)
ax.set_xlabel(r"$t_{det}~[\omega_p^{-1}]$")
ax.set_ylabel(r"$\phi\;$[mrad]")
```
